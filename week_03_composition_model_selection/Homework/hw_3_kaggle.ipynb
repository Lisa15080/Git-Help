## Применение линейных моделей (3 балла)

1) Обработайте данные для того, чтобы к ним можно было применить LogisticRegression. Т.е. отнормируйте числовые признаки, а категориальные закодируйте с помощью one-hot-encoding'а.

2) С помощью кроссвалидации или разделения на train/valid выборку протестируйте разные значения гиперпараметра C и выберите лучший (можно тестировать С=100, 10, 1, 0.1, 0.01, 0.001) по метрике ROC-AUC.

(более сложный вариант) Если вы будете использовать кроссвалидацию, то преобразования данных и LogisticRegression нужно соединить в один Pipeline с помощью make_pipeline, как это делалось во втором семинаре. Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.compose import ColumnTransformer
X_all = data.drop(columns=['Churn'])
y_all = data['Churn']

X_train, X_valid, y_train, y_valid = train_test_split(
    X_all, y_all, test_size=0.2, random_state=42, stratify=y_all
)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols)
    ],
    remainder='drop'
)

X_train_proc = preprocessor.fit_transform(X_train)
X_valid_proc = preprocessor.transform(X_valid)

C_values = [100, 10, 1, 0.1, 0.01, 0.001]

best_C = None
best_roc_auc = -1.0

print(f"{'C':<8} | {'ROC-AUC':<10}")
print("-" * 20)

for C in C_values:
    model = LogisticRegression(C=C, class_weight='balanced', random_state=42, max_iter=1000)
    model.fit(X_train_proc, y_train)
    y_valid_proba = model.predict_proba(X_valid_proc)[:, 1]
    current_score = roc_auc_score(y_valid, y_valid_proba)
    print(f"{C:<8.4f} | {current_score:<10.4f}")
    if current_score > best_roc_auc:
        best_roc_auc = current_score
        best_C = C

print(f"\n Лучший параметр: C = {best_C} (ROC-AUC = {best_roc_auc:.4f})")

final_model = LogisticRegression(C=best_C, class_weight='balanced', random_state=42, max_iter=1000)
final_model.fit(X_train_proc, y_train)

```

## Применение градиентного бустинга (2 балла)

Если вы хотите получить баллы за точный ответ, то стоит попробовать градиентный бустинг. Часто градиентный бустинг с параметрами по умолчанию даст вам 80% результата за 0% усилий.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

2) Протестируйте разные занчения параметров количества деревьев и learning_rate'а и выберите лучшую по метрике ROC-AUC комбинацию.

(Дополнительно) Есть некоторые сложности с тем, чтобы использовать CatBoostClassifier вместе с GridSearchCV, поэтому мы не просим использовать кроссвалидацию. Но можете попробовать)

```python
from catboost import CatBoostClassifier

model_default = CatBoostClassifier(random_state=42, verbose=0)
model_default.fit(X_train, y_train, cat_features=cat_cols, eval_set=(X_valid, y_valid))
roc_default = roc_auc_score(y_valid, model_default.predict_proba(X_valid)[:, 1])
print(f"ROC-AUC (default): {roc_default:.4f}\n")

iters_list = [100, 300, 500, 1000]
lrs = [0.01, 0.05, 0.1, 0.2]

best_roc, best_params, best_model = -1.0, {}, None

print(f"{'Iter':<6} | {'LR':<6} | {'ROC-AUC'}")
print("-" * 28)

for it in iters_list:
    for lr in lrs:
        model = CatBoostClassifier(
            iterations=it,
            learning_rate=lr,
            random_state=42,
            verbose=0,
            early_stopping_rounds=30
        )

        model.fit(X_train, y_train, cat_features=cat_cols, eval_set=(X_valid, y_valid))

        score = roc_auc_score(y_valid, model.predict_proba(X_valid)[:, 1])
        print(f"{it:<6} | {lr:<6.2f} | {score:.4f}")

        if score > best_roc:
            best_roc, best_params, best_model = score, {'iterations': it, 'learning_rate': lr}, model

print(f"\nЛучшая комбинация: {best_params}")
print(f"Лучший ROC-AUC: {best_roc:.4f}")
```

Выпишите, какое лучшее качество и с какими параметрами вам удалось получить

Лучший ROC-AUC, полученный с помощью градиентного бустинга, = 0.8566. Параметры: iterations=100, leraning_rate=0.1
